```python
x=np.arange(10)
y1=x**2
y2=2*x+3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1,color='red')
plt.show()
plt.plot(x,y2,color="green")
plt.show()
```

```python
plt.style.use("bmh")
plt.plot(x,y1,color='red',label="apple",marker='o')
#plt.show()
plt.plot(x,y2,color="green",label="kiwi",linestyle='dashed',marker='*')
plt.xlabel("time")
plt.ylabel("price")
plt.title("prices of fruits over time")
plt.legend()
plt.show()
```

Scatter Plots

```python
plt.scatter(x,y1,color='red',label="apple",marker='^')
#plt.show()
plt.scatter(x,y2,color="green",label="kiwi",linestyle='dashed',marker='*')
plt.xlabel("time")
plt.ylabel("price")
plt.title("Scatter Plots-prices of fruits over time")
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x_coardinates=np.array([0,1,2])*2
plt.bar(x_coardinates,[10,20,15],width=0.5,label="Current year",tick_label=["Gold","Silver","Platinum"],color="green")
plt.bar(x_coardinates+0.5,[20,10,12],width=0.5,label="Next year",color="red")
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.xlim(-1,5)
plt.legend()
plt.show()
```

```python
plt.style.use("seaborn")
subjects="Maths","chem","physics","English"
weightage=[20,10,17,20]
plt.pie(weightage,labels=subjects,explode=(0,0,0.5,0),autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
#HISTOGRAM
```

```python
#xsn=np.random.randn?
Xsn=np.random.randn(100)
sigma=5
u=70
X=np.round(Xsn*sigma+u)
X2=np.round(Xsn*15+40)
print(X,X2)
```

```python
plt.hist(X,label="physics")
plt.xlabel("marks range")
plt.ylabel("frequency")
plt.hist(X2,label="Maths")

plt.title("Histogram")
plt.show()
```
